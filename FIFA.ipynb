## **Using Mean Values in NaN**

```python
# names = df.short_name.tolist() # saving names for later

# df = df.drop(['short_name'], axis = 1) # drop the short_name column 

# keeping names of player in a different df, this will be used later 
names = df_mean.short_name 

# dropping those columns which might induce redundancy
# overall & potential is a function of all other attributes like physical, defending, pace, dribbling etc. so that's why dropping
# value_eur & wage_eur don't generally contribute to the technical attributes
to_drop = ['short_name', 'overall', 'potential', 'value_eur', 'wage_eur', 'age',
       'height_cm', 'weight_kg']
attributes = df_mean.drop(to_drop, axis = 1 )

attributes.head()
```

Using PCA to reduce dimensionality

```python
from sklearn.decomposition import PCA

# 2D PCA for the plot
pca = PCA(n_components = 2)
reduced_mean = pd.DataFrame(pca.fit_transform(attributes))
```

Using K-means clustering

```python
from sklearn.cluster import KMeans

# specify the number of clusters
kmeans = KMeans(n_clusters=10)

# fit the input data
kmeans = kmeans.fit(reduced_mean)

# get the cluster labels
labels = kmeans.predict(reduced_mean)

# centroid values
centroid = kmeans.cluster_centers_

# cluster values
clusters = kmeans.labels_.tolist()
```

Creating new DF using player names and above formed clusters

```python
reduced_mean['cluster'] = clusters
reduced_mean['name'] = names
reduced_mean.columns = ['x', 'y', 'cluster', 'name']
reduced_mean.head()
```

## **Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set(style="white")

ax = sns.lmplot(x="x", y="y", hue='cluster', data = reduced_mean, legend=False,
                   fit_reg=False, height = 15, scatter_kws={"s": 250})

texts = []
for x, y, s in zip(reduced_mean.x, reduced_mean.y, reduced_mean.name):
    texts.append(plt.text(x, y, s))

#ax.set(ylim=(-2, 2))
plt.tick_params(labelsize=15)
plt.xlabel("PC 1", fontsize = 20)
plt.ylabel("PC 2", fontsize = 20)

plt.show()
```
